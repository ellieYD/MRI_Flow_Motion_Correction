```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import os
```

```python
directory = '/Users/ellie/Desktop/FYP/Data'  # Replace with your target directory
file_name = 'avg_phase_mod.csv' #phase modulation results file
csv_file_path = os.path.join(directory, file_name)

data = pd.read_csv(csv_file_path, index_col=0, sep=',')

array_data = data.to_numpy()
```

```python
fig, ax = plt.subplots(figsize=(8, 6),dpi = 1000)

cax = ax.imshow(np.transpose(array_data[21:40]), cmap='viridis', aspect='auto')

ax.set_xticks(np.arange(21,41)-21)
ax.set_xticklabels(np.arange(21,41))

plt.xlabel('TR Spacing')
plt.ylabel('PE Line')
plt.title('Phase Modulation of Avergaed K-Space')
plt.colorbar(cax)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(array_data[0,:], label='1')
plt.plot(array_data[5,:], label='6')
plt.plot(array_data[15,:], label='16')
plt.plot(array_data[31,:], label='32')
plt.plot(array_data[127,:], label='128')
plt.legend(title='TR spacing', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
plot_indices = np.array([0,5,15,31,127])

plt.figure(figsize=(8, 6),dpi = 1000)
plt.imshow(array_data[plot_indices,:])
plt.colorbar()  
plt.title('')
plt.xlabel('TR Spacing')
plt.ylabel('PE Line')
plt.show()
```
